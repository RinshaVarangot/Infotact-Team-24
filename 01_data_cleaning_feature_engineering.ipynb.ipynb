```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df['Country'] = df['Country'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.drop(['Agent','Company'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df['total_guests'] = df['Adults'] + df['Children'] + df['Babies']
```

```python
df['total_stay'] = df['StaysInWeekendNights'] + df['StaysInWeekNights']
```

```python
df['revenue'] = df['ADR'] * df['total_stay']
```

```python
month_map = {
"January":1,"February":2,"March":3,"April":4,"May":5,"June":6,
"July":7,"August":8,"September":9,"October":10,"November":11,"December":12
}

df["arrival_month_number"] = df["ArrivalDateMonth"].map(month_map)
```

```python
df.to_csv("cleaned_hotel_bookings.csv", index=False)
```

```python
sns.countplot(x="IsCanceled", data=df)
plt.title("Cancellation Distribution")
plt.show()
```
