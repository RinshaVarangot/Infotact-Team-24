## Missing Value Analysis

```python
missing=df.isnull().sum().sort_values(ascending=False)
missing_percent=(missing/len(df))*100

pd.DataFrame({
    "Missing Count":missing,
    "Missing %":missing_percent
})
```

## Univeariate Analysis-Numerical

```python
num_cols=['iscanceled','leadtime','isrepeatedguest','adr']                          
              
 
for col in num_cols:
    plt.figure()
    sns.histplot(df[col],kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

1.  Cancellation Rate is High
   
Overall cancellation rate: ~23.5%

Nearly 1 in 4 bookings gets canceled

👉 Impact: Revenue leakage + poor occupancy planning

2. Lead Time Strongly Impacts Cancellation
   
Short bookings (≤ 7 days): ~6% cancellation

Medium (30–90 days): ~30% cancellation

Long (90–365 days): ~32–34% cancellation

👉 Insight:

Earlier bookings = higher cancellation risk

Customers booking far in advance are less committed

👉 Action:

Introduce partial prepayment / stricter policies for long lead bookings

3.  Repeat Guests Are Low
   
Very small % of repeated guests

👉 Insight:

Business is acquisition-heavy, not retention-driven

👉 Action:

Focus on:

Membership programs

Personalized offers

Post-stay engagement

4. Revenue Insight (ADR – Average Daily Rate)
   
Wide variation in ADR (pricing inconsistency)

👉 Likely:

Dynamic pricing not optimized

Different segments paying very different prices

👉 Action:

Improve revenue management strategy
Use demand-based pricing mode

## Univariate Analysis - Categorical

```python

cat_cols = ['marketsegment','deposittype']

for col in cat_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f"Top Categories in {col}")
    plt.show()
```

1.  Market Segment Performance

Highest cancellation rates:

Online TA (Travel Agents): ~34% 

Groups: ~19%

Direct bookings: ~13%

👉 Insight:

Online platforms bring volume but low reliability

👉 Action:

Offer incentives for direct bookings

Re-evaluate OTA dependency (commission vs risk)

2. 💳Deposit Type = Biggest Cancellation Driver

Non-refundable: ~84% cancellation  (very unusual pattern)

No deposit: ~23%

Refundable: ~17%

👉 Insight:

Something is wrong or misleading here:

Either data labeling issue

OR customers book multiple non-refundable options and cancel strategically

👉 Action:

Audit booking policy logic immediately

Validate pricing & refund rules



## Outlier Detection

```python
for col in ['adr']:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

## Correlation Analysis

```python
plt.figure(figsize=(18,12))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
